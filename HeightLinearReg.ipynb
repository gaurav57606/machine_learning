```python
df['Gender'] = df['Gender'].replace({'M':0,'F':1})
X = df[['Gender', 'Father', 'Mother']]
y = df['Height']
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
print(lr.intercept_)
print(lr.coef_)
```

```python
y_pred = lr.predict(X)
```

```python
residual = y-y_pred
```

```python
residual.shape, y.shape
```

```python
plt.scatter( x = y_pred, y = residual);
```
